```python
# Drop null values
df_harvest.dropna(inplace=True)

df_harvest["year"] = pd.to_datetime(df_harvest["year"])


# Filter data for Minas Gerais
minas_harvest = df_harvest[df_harvest["subdivision"] == "Minas Gerais"]

minas_harvest.head()
```

```python
# Harvest Trends Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df[df["subdivision"] == "Minas Gerais"], x="year", y="million_60kgs_bag", label="Minas Gerais", marker="o")
sns.lineplot(data=df, x="year", y="million_60kgs_bag", hue="subdivision", alpha=0.3, marker="o")
plt.title("Harvest Trends Over Time (Minas Gerais vs. Other Regions)")
plt.xlabel("Year")
plt.ylabel("Production (Million 60kg Bags)")
plt.legend()
plt.show()
```

```python
"""""""""
This line plot shows the coffee production trends in Minas Gerais between the years of 2002 and 2023.
Minad Gerais is being compared to the other regions to determine if the data shows consistency. Having other regions
to compare this data to, creates a visualization of the production trends in Minas Gerais. 
We can see that Minas Gerais contributes to most of Brazil's coffee production, while Parana contributes to the least.
This is due to the fact that Minas Gerais has the most suitable climate and geographical features for coffee production.

"""""""""
```

```python
#Unemployment Trends in Coffee-Producing Regions
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="year", y="avg_unemp_perc", hue="subdivision", marker="o")
plt.title("Unemployment Trends in Coffee-Producing Regions")
plt.xlabel("Year")
plt.ylabel("Average Unemployment Percentage")
plt.legend()
plt.show()

```

```python
"""""""""""
The line plot shows how the average unemployment percentage has changed in Minas Gerais and other regions between 2002 
and 2023. The data shows that the unemployment rate in Minas Gerais has the highest average unemployment rate compared 
to Parana. And has less unemployment compared to Sao Paulo and Espirito Santo. A decrease in coffee production could
result into the decrease of availabe for emplyers and employees. Therefore, the decrease in coffee production will 
increase the unemployment rate in the coffee-producing regions. 

"""""""""
```

```python

df_weather = pd.read_csv("/Users/sa3/Desktop/TLAB3/csv/weather_data.csv")
df_weather.dropna(inplace=True)
df_weather["time"] = pd.to_datetime(df_weather["time"])
df_weather.set_index("time", inplace=True)
print(df_weather.head())

# Temperature Trend in Minas Gerais
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_weather, x="time", y="temperature_2m", marker="o")
plt.title("Weather Outcomes in Minas Gerais (Temperature Trend)")
plt.xlabel("Year")
plt.ylabel("Average Temperature (°C)")
plt.show()
```

```python
""""""""""
This line plot shows the average temperature trends in Minas Gerais over the years. Between April 2022-July 2022 and April 2023-July 2023, the 
temperature was at its lowest. This can be due to seasonal changes such as when the temperature naturally decreases in winter 
which causes less direct sunlight. This will result in a decrease in coffee production. The temperature is at its highest between Octiober 2022 and
January 2023. This is when the temperature is at its highest and the sun is at its brightest. This will result in an increase in coffee production.

"""""""""
```
